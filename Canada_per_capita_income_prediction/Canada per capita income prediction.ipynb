```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
df= pd.read_csv("canada_per_capita_income.csv")
```

```python
df.head()
```

```python
%matplotlib inline
plt.xlabel('per capita income (US$)',fontsize=20)
plt.ylabel('year',fontsize=20)
plt.scatter(df['per capita income (US$)'],df.year,color='red',marker='+')
plt.title("Year vs per capita income (US$")
```

```python
X = df[['year']]  # Feature: Year
y = df['per capita income (US$)']  # Target variable


reg = linear_model.LinearRegression()
reg.fit(X,y)
```

```python
reg.predict([[2025]])
```

```python
reg.score(X,y)
```
